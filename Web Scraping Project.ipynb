# Web Scraping

<h3> Necessay Libraries </h3>

```python
import requests
import bs4
import pandas as pd
```

<h3> Scrape from the base page </h3>

```python
url = "http://books.toscrape.com/index.html"
book = requests.get(url)
bk = bs4.BeautifulSoup(book.text,'lxml')
lst_url = ["http://books.toscrape.com/"+(str(i).split(" ")[1].split("=")[-1]).replace('"',"") for i in bk.select('h3')]
lst_url

lst_df = []
for m in lst_url:
    
    url1 = m
    books = requests.get(url1)
    bks1 = bs4.BeautifulSoup(books.text,'lxml')

    lst_val = [j[1:] if i in(2,3,4) else j for i,j in enumerate([i.getText() for i in bks1.select('td')])]
    
    title = bks1.select('h1')[0].getText() ##title
    lst_val.append(title)
    rat = [str(i) for i in bks1.select('p')]
    description = [i for i in rat if "<p>" in i][0].replace("<p>","").replace("</p>","") ##book description
    lst_val.append(description)
    
    rating =[i for i in rat if "star-rating" in i][0].split(">\n")[0][9:].replace('"',"")
    lst_val.append(rating)
    col_names =["UPC","PRODUCT_TYPE","PRICE(exlusive_tax)","PRICE(inclusive_tax)","TAX","AVAILABILITY","REVIEWS","TITLE","PRODUCT_DESCRIPTION","RATINGS"]

    df = pd.DataFrame(data= {x:[y] for x,y in zip(col_names, lst_val)}).loc[:,["TITLE","PRODUCT_DESCRIPTION","UPC","PRODUCT_TYPE","RATINGS","REVIEWS","AVAILABILITY","PRICE(exlusive_tax)","PRICE(inclusive_tax)","TAX",]]
    lst_df.append(df)

df_base = pd.concat(lst_df)
```
